```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df=pd.read_excel("CC (1).xlsx")
df.head()
```

```python
df.isnull().sum() #To check for any missing values.
```

```python
df.describe(include='all') #To describe both categorical and continuous variables..
```

```python
df.drop_duplicates(subset=['Application id']) #To drop any duplicate rows(if there are..)
```

```python
for i in range(0, len(df)):#Imputing
    if ((df.loc[i,'tdecision']=='Decline')|(df.loc[i,'tdecision']=='Pending')):
        df.loc[i,('Booking','Booking_Amt')] = np.nan
```

```python
df.head()
```

```python
df['Net_Income']=df['Salary']+df['ExDebt'] #feature generation by addition of existing features
df[['Application id','Net_Income']]
```

```python
for i in range(0, len(df)):#Imputing
    if (df.loc[i,'Net_Income']<=0):
        df.loc[i,('Booking','Booking_Amt')] = np.nan
df.head()
```
